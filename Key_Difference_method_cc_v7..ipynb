```python
import string
import numpy as np
import json
import pandas as pd
import requests
```

```python
def FindKeyDifference(file_link):
    #Loading the similarity JSON file
    f = open(file_link)
    response = json.load(f)
    sim_1 = []
    for i in range(len(response)):
        for value in response[i].values():
            #print(value)
            sim_1.append(value)
    li=[]
    li_4=[]

    #Function to find difference words
    def gth(strings22):
        if (len(strings22)<=2):
            A=strings22[0].split()
            B=strings22[1].split()
            #print(A,B)
            x=[]
            y=[]
            z=[]
            for i in A:
                if i not in B:
                    x.append(i)
            for i in B:
                if i not in A:
                    y.append(i)
            #x=set(x)
            #y=set(y)
            #print(x, y)
            z.append(x)
            z.append(y)
            return z
        else:
            uncommon_words = []

            for i in range(len(strings22)):
                for j in range(i + 1, len(strings22)):
                    words1 = strings22[i].split()
                    words2 = strings22[j].split()
                    uncommon_in_1 = [word for word in words1 if word not in words2]
                    uncommon_in_2 = [word for word in words2 if word not in words1]
                    uncommon_words.append((strings22[i], strings22[j], uncommon_in_1, uncommon_in_2))
            result = uncommon_words
            #print(result)
            if len(result) > 0:
                str_3 = []
                str_4 = []
                for diff in result:
                    str1, str2, uncommon_in_1, uncommon_in_2 = diff
                    str_3.append(str1)
                    str_3.append(str2)
                    str_4.append(uncommon_in_1)
                    str_4.append(uncommon_in_2)
            dict_5 = {}
            for i in range(len(str_3)):
                for j in range(i+1, len(str_3)):
                    if str_3[i] == str_3[j]:
                        dict_5.update({str_3[i]:str_4[i]})
                        #dict_5.append(str_4[i])
            #print(dict_5)
            z=[]
            for n in dict_5:
                z.append(dict_5[n])
            return z
    #Main logic to store and manuplate JSON and Differences
    for i in range(10):
        dict_2 = {}
        li_3= []
        if len(sim_1[i]) == 0:
            i = i+1
        else:
            #dict_1 = {}
            li_2 = []
            li_8 = []
            for k in range(len(sim_1[i])):
                dict_1 ={}
                #print(sim_1[i][k])
                url_1 = sim_1[i][k]
                url_2 = url_1+'.json'
                #print(url_2)
                r1 = requests.get(url_2)
                #r1.raise_for_status()  # raises exception when not a 2xx response
                #if (r1.status_code != 204) or (r1.status_code != 430):
                #Load JSON into variable
                r1_json = json.loads(r1.text)
                #print(r1_json)
                products_json = r1_json['product']
                #print(products_json)
                if products_json:
                    for pattern in products_json:
                        if pattern == 'title':
                            #print(pattern, '==', products_json[pattern])
                            dict_1.update({"product_link":url_1})
                            dict_1.update({"product_title":products_json[pattern]})
                            li_2.append(products_json[pattern])
                            li_3.append(dict_1)
                            li_8.append(dict_1)
            #Get list and differences
            li_7 = gth(strings22=li_2)
            #Update these differences into the dictionary 
            for i in range(len(li_8)):
                #print(li_8[i])
                li_8[i].update({"product_differences":li_7[i]})
                #print(li_8[i])
                #print('--------------')
            #print(li_8) 
            #print(li_8[1])             
            #print(li_2)
            #print(dict_1)
            #li_3.append(dict_1)
            li_3 = li_8
            li_4.append(li_2)
        #print(li_3)
        #Update dictornary to list
        li.append(li_3)
    return li
```
